```python
import pandas as pd
import numpy as np
seasons = {}
for year in range(1980, 2017):
    sheet = str(year)
    df = pd.read_excel("MLB_batting_data_1980_2018_19_10_02.xlsx", sheet_name=sheet)
    seasons[sheet] = df[['Name', 'H', 'AB', 'BA', 'OPS']]
seasons
```

```python
differences = {}

for year in range(1980, 2016):
    s_first = seasons[str(year)]
    s_first['Name'] = s_first['Name'].apply(lambda x: x.strip("*").strip("#").split('\\')[0])
    s_first = s_first.set_index('Name')
    s_second = seasons[str(year + 1)]
    s_second['Name'] = s_second['Name'].apply(lambda x: x.strip("*").strip("#").split('\\')[0])
    s_second = s_second.set_index('Name')
    intersection = list(set(s_first.index).intersection(set(s_second.index)))
    s_first = s_first[(s_first['H'].notna()) & (s_first['AB'].notna())]
    s_second = s_second[(s_second['H'].notna()) & (s_second['AB'].notna())]
    
    first_batting_hits = 0.0
    first_batting_abs = 0.0
    second_batting_hits = 0.0
    second_batting_abs = 0.0
    
    total_counted = 0
    for player in intersection:
        try:
            if type(s_first.at[player, 'H']) != np.int64 or type(s_second.at[player, 'H']) != np.int64 or type(s_first.at[player, 'AB']) != np.int64 or type(s_second.at[player, 'AB']) != np.int64:
                continue
        except:
            continue
        try:
            first_batting_hits = first_batting_hits + s_first.at[player, 'H']
            second_batting_hits = second_batting_hits + s_second.at[player, 'H']
            first_batting_abs = first_batting_abs + s_first.at[player, 'AB']
            second_batting_abs = second_batting_abs + s_second.at[player, 'AB']
            total_counted += 1
        except:
            continue
               
    differences[str(year +1) + "-" + str(year)] = (second_batting_hits)/(second_batting_abs) - (first_batting_hits)/(first_batting_abs)
```

```python
differences
```

```python
years = ['2000-1999', '2001-2000', '2002-2001', '2003-2002', '2004-2003', '2005-2004', '2006-2005', '2007-2006',
        '2008-2007', '2009-2008', '2010-2009', '2011-2010',]
```

```python
diff = 0
for year in years:
    diff += differences[year]
diff
```
