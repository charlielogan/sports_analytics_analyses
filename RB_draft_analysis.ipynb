```python
import pandas as pd
```

```python
df = pd.read_csv("2017combine.csv")
df = df[df['Pos'] == 'RB']
```

```python
df['WA40'] = (df['Wt']*200)/(df['40yd']**4)
```

```python
df['Player'] = df['Player'].str.split("\\", n=0).str[0]
df = df.sort_values('WA40', ascending=False)
df = df[df['WA40'].notna()]
df["Rank"] = range(1, 28)
df['Pick'] = df["Drafted (tm/rnd/yr)"].str.split("/").str[2].str.split(" pick").str[0].str[0:-2]
```

```python
df.set_index('Player')[["Rank", "Pick", "WA40", "School", 'Ht', "Wt", '40yd', "Drafted (tm/rnd/yr)"]]
```
