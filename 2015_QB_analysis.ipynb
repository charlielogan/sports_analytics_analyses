```python
import pandas as pd
df = pd.read_csv('2015QBs.csv')
df
```

```python
df = df[:32]
df['Player'] = df['Player'].apply(lambda x: x.split('*')[0])
df['Player'] = df['Player'].apply(lambda x: x.split('\\')[0])
df['Player'] = df['Player'].apply(lambda x: x.split()[0] + " " + x.split()[1])
```

```python
def check_min_max(calc):
    if calc > 2.375:
        return 2.375
    elif calc < 0:
        return 0
    else:
        return calc

def find_nfl_rating(row):
    a = check_min_max(((row['Cmp']/row['Att']) - .3) * 5)
    b = check_min_max((row['Y/A'] - 3) * .25)
    c = check_min_max(((row['TD']/row['Att']) * 20))
    d = check_min_max((2.375 - ((row['Int']/row['Att']) *25)))
    return ((a+b+c+d)/6)*100

def find_wages_of_wins(row):
    return (row['Yds'] - row['Yds.1']) - (3*row['Att']) - (30*row['Int'])

def find_burke_rating(row):
    return (1.54337 * ((row['Yds'] - row['Yds.1']) / (row['Att'] + row['Sk']))) - (50.0957 * (row['Int']/row['Att']))

df['nfl_rating'] = df.apply(find_nfl_rating, axis=1)
df['wages_rating'] = df.apply(find_wages_of_wins, axis=1)
df['burke_rating'] = df.apply(find_burke_rating, axis=1)
```

```python
df1 = df.loc[: , ['Rk', 'Player', "Tm", 'Age', 'G', 'Yds', 'Yds.1', 'Cmp', 'Att', 'Sk', 'TD', "Int", 'nfl_rating', 'wages_rating', 'burke_rating']].set_index('Rk')
df1
```

```python
dfn = df1.sort_values(by='nfl_rating', ascending=False).head(10)
dfn.index = range(1, 11)
dfn
```
